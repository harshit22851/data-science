```python
import zipfile

# specify the path of the zip file
zip_path = "EyeT.zip"

# create a ZipFile object and extract all files
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("EyeT")
```

```python
import pandas as pd
import os

# Define the directory containing the CSV files
directory = 'EyeT/EyeT/'

df_list = [] 


# Loop through each file in the directory
for filename in os.listdir(directory):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(os.path.join(directory, filename))
        
        # Perform the necessary operations
        df_numeric = df.select_dtypes(include=['float64', 'int64'])
        df_mean = df_numeric.mean()
        df_mean = df_mean.to_frame().T
        df = df_mean.iloc[:, 0:]
        df_other = pd.read_csv(os.path.join(directory, filename), usecols=["Participant name"]).head(1)
        df_concat = pd.concat([df, df_other], axis=1)
        df_list.append(df_concat)

df=pd.concat(df_list)
```

```python
df.to_csv('dataset2.csv')
```

```python
import pandas as pd


#Sorting
df = df.sort_values(by=['Participant name'])
# get the name of the last column
last_col_name = df.columns[-1]

# create a list of the column names with the last column name removed
new_col_order = [last_col_name] + list(df.columns[:-1])

# rearrange the columns in the dataframe
df = df[new_col_order]

# write the merged dataframe to a new CSV file
df.to_csv('Final1.csv', index=False)
```
